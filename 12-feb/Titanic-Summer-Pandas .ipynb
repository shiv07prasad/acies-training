### Indexing using iloc

```python
summer = pd.read_csv('summer.csv')
```

```python
summer.head()
```

```python
summer.tail()
```

```python
summer.info() #missing value in country
```

```python
summer.iloc[0] #row 0
```

```python
type(summer.iloc[0])
```

```python
summer.iloc[1]
```

```python
summer.iloc[-1]
```

```python
summer.iloc[:5] #first five rows
```

```python
summer.iloc[-5:] #last five rows
```

```python
summer.iloc[455:461]
```

```python
summer.iloc[[2,56,153,752]]  #pass index positions using lists
```

```python
summer.iloc[0,4]
```

```python
summer.iloc[0,0:3]
```

```python
summer.iloc[0,[2,4,6,8]]
```

```python
summer.iloc[34:40,[2,4,6,8]]
```

```python
summer.iloc[:,4]
```

```python
summer.iloc[:,4].equals(summer.Athlete)
```

### Label based indexing using .loc

```python
summer = pd.read_csv('summer.csv',index_col = 'Athlete')
```

```python
summer.head()
```

```python
summer.iloc[2]
```

```python
summer.loc['DRIVAS, Dimitrios']
```

```python
summer.loc['PHELPS, Michael']
```

```python
summer.loc['PHELPS, Michael'].iloc[0]
```

### Slicing rows and columns with loc

```python
summer.loc['PHELPS, Michael', 'Medal']
```

```python
summer.loc['PHELPS, Michael', ['Medal','Event']]
```

```python
summer.loc[['PHELPS, Michael','LEWIS, Carl'], ['Medal','Event']]
```

```python
summer.loc[:'CHASAPIS, Spiridon']
```
