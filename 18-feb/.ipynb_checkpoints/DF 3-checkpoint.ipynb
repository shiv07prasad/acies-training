### apply() , map() and applymap()

```python
import pandas as pd
```

```python
sales = pd.read_csv('sales.csv')
```

```python
sales
```

```python
sales = pd.read_csv('sales.csv',index_col = 0)
```

```python
sales
```

```python
sales.info()
```

```python
sales.min(axis = 0)
```

```python
sales.min(axis = 1)
```

```python
def range(series):
    return series.max() - series.min()
```

```python
sales.apply(range,axis = 0)
```

```python
sales.apply(range,axis = 1)
```

```python
sales.apply(lambda x: x.max() - x.min(),axis = 1)
```

```python
summer = pd.read_csv('summer.csv')
```

```python
summer.head()
```

```python
summer.Athlete.apply(lambda x: x[0])
```

```python
summer.Athlete.map(lambda x: x[0])
```

```python
summer.iloc[:,1:3].applymap(lambda x: x[0])
```

```python
sales.applymap(lambda x: 0.4*x-5)
```

```python
sales*0.4 - 5
```

### heirarchical indexing 

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.set_index(['pclass','sex'],inplace = True)
```

```python
titanic.sort_index(ascending = [True,True], inplace = True)
```

```python
titanic
```

```python
titanic.swaplevel()
```

```python
titanic.reset_index(inplace = True)
```

```python
titanic
```

### heirarchical index

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic = titanic.set_index(['pclass','sex']).sort_index(ascending = True)
```

```python
titanic
```

```python
titanic.loc[1]
```

```python
titanic.loc[[1,2]]
```

```python
titanic.loc[:2]
```

```python
titanic.loc[1,'female']
```

```python
titanic.loc[1,'female','age']
```

```python
titanic.loc[(1,'female')]
```

```python
titanic.loc[(1,'female'),'age']
```

```python
titanic.loc[([1,2],'female'),['age','fare']]
```

```python
titanic.loc[(slice(None),slice('female')),:]
```
