### View vs copy

```python
titanic = pd.read_csv('titanic.csv')
```

```python
age = titanic.age
```

```python
age.head()
```

```python
age._is_view
```

```python
age._is_copy is None
```

```python
age[1] = 40
```

```python
age.head()
```

```python
titanic.head()
```

### slicing a df and creating a copy

```python
df_baby = titanic[titanic.age < 1]
```

```python
df_baby._is_view
```

```python
df_baby._is_copy is None
```

```python
df_baby.age = 1
```

```python
df_baby
```

```python
titanic.loc[index_babies]
```

### make a copy

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
age = titanic.age.copy()
```

```python
age[1] = 40
```

```python
age.head()
```

```python
titanic.head()
```

```python
baby_ages = titanic.loc[titanic.age < 1,['age','sex']].copy()
```
