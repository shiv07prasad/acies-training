### Sorting DataFrames with sort_index() and sort_values()

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
titanic.age.sort_values()
```

```python
titanic.sort_values('age')
```

```python
titanic.head()
```

```python
titanic.sort_values('age',axis = 0,ascending = False, inplace = False)
```

```python
titanic.head()
```

```python
titanic.sort_values(['pclass','sex','age'],ascending = [True, False, True])
```

```python
titanic.head()
```

```python
titanic.sort_index(ascending = True,inplace = True)
```

```python
titanic.head()
```

### Ranking DataFrames with rank()

```python
sales = pd.Series([15,32,45,21,55,15,0], index = ['Mo','Tue','Wed','Thu','Fri','Sat','Sun'])
```

```python
sales
```

```python
sales.sort_values(ascending = False)
```

```python
sales.rank()
```

```python
sales.rank().sort_values(ascending = False)
```

```python
sales.rank(ascending = False)
```

```python
sales.rank()
```

```python
sales.rank(ascending = False, method = 'min')
```

```python
sales.rank(ascending = False, method = 'min',pct = True).sort_values()
```

```python
titanic.fare.rank(ascending = True)
```

```python
titanic['fare_rank']= titanic.fare.rank(ascending = False, method = 'min')
```

```python
titanic.head()
```

### nunique() , nlargest(), nsmallest() with DF

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.age.unique()
```

```python
titanic.nunique(axis = 0)
```

```python
titanic.nunique(dropna  = False)
```

```python
titanic.nlargest(n = 5, columns = 'fare')
```

```python
titanic.sort_values('fare',ascending = False).head()
```

```python
titanic.nsmallest(n = 5,columns = 'age')
```

```python
titanic.loc[titanic.age.idxmin()]
```

### The agg() method

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.describe()
```

```python
titanic.agg('mean')
```

```python
titanic.agg(['mean','std'])
```

```python
titanic.agg(['mean','std','min','max','median'])
```

```python
titanic.agg({'survived':'mean','age':['min',max]})
```
