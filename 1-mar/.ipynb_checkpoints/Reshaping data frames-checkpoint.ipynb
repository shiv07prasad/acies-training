```python
import pandas as pd
```

```python
titanic.head()
```

```python
titanic.T
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic = titanic.T
```

```python
titanic.head()
```

```python
titanic.survived.mean()
```

```python
titanic = pd.read_csv('titanic.csv')
```

## Pivoting dataframes with pivot()

```python
import pandas as pd
```

```python
table1 = pd.read_csv('table1.csv')
```

```python
table1.head()
```

```python
table1.info()
```

```python
table1.shape
```

```python
table1.pivot(index = 'Country',columns = 'Medal',values = 'Count')
```

```python
table_piv = table1.pivot(index = 'Country',columns = 'Medal',values = 'Count').fillna(0)
```

```python
table_piv.info()
```

```python
table1.set_index(['Country','Medal']).unstack(fill_value = 0)
```

```python
table_piv
```

## pivot_table() (Part 1)

```python
table2 = pd.read_csv('table2.csv')
```

```python
table2.head()
```

```python
table2.pivot_table(index = 'Country',columns = 'Medal',values = 'Count',aggfunc = 'sum',fill_value = 0, margins = True)
```

```python
table2.groupby(['Country','Medal']).Count.sum().unstack(fill_value = 0 )
```

```python
agg = table2.groupby(['Country','Medal']).Count.sum().unstack(fill_value = 0 )
```

```python
agg['All'] = agg.sum(axis = 1)
```

```python
agg
```

## pd.crosstab()

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.info()
```

```python
pd.crosstab(titanic.sex, titanic.pclass,margins = True) #frequency table
```

```python
titanic.groupby(['sex','pclass']).pclass.count().unstack()
```

```python
pd.crosstab(titanic.sex, titanic.pclass, margins = True, normalize = True )
```

```python
titanic.groupby(['sex','pclass']).pclass.count().unstack()
```

```python
pd.crosstab(index = titanic.sex, columns = titanic.pclass, values = titanic.age, aggfunc = 'mean')
```

```python
titanic.pivot_table(index = 'sex', columns = 'pclass', values = 'age',aggfunc = 'mean')
```

```python
titanic.groupby(['sex','pclass']).age.mean().unstack()
```

```python
pd.crosstab(index = titanic.sex, columns = titanic.pclass, values = titanic.fare, aggfunc = 'sum',margins = True, normalize =True)
```

```python
summer = pd.read_csv('summer.csv')
summer
```

```python
pd.crosstab(index = [summer.Year,summer.Country],columns = summer.Medal, aggfunc = 'count',values = summer.Athlete, margins = True).fillna(0)
```

```python
summer.pivot_table(index = ['Year','Country'],columns = 'Medal',aggfunc = 'count',values = 'Athlete',fill_value = 0,margins = True)
```

```python
summer.groupby(['Year','Country','Medal']).Medal.count().unstack(fill_value = 0)
```
