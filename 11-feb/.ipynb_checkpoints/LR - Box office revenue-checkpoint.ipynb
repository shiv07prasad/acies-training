## Linear Regression

### Importing the libraries

```python
import pandas as pd
```

```python
data = pd.read_csv('3.2 cost_revenue_clean.csv.csv')
data.head()
```

```python
data.describe()
```

### Splitting feature and target variable

```python
X = data['production_budget_usd'].values.reshape(-1,1)
y = data['worldwide_gross_usd'].values.reshape(-1,1)
```

### Visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize = (12,10))
plt.scatter( X, y, alpha = 0.3)
plt.title('Production Budget v Worldwide gross')
plt.xlabel('Production budget')
plt.ylabel('Worldwide gross')
plt.xlim(left = 0)
plt.ylim(bottom = 0)
plt.show()
```

### Implementing the algorithm

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.get_params()
```

```python
#slope co-efficient
reg.coef_   
```

```python
# intercept
reg.intercept_
```

```python
plt.figure(figsize = (12,10))
plt.scatter( X, y, alpha = 0.3)
plt.plot(X, reg.predict(X), c = 'red')


plt.title('Production Budget v Worldwide gross')
plt.xlabel('Production budget')
plt.ylabel('Worldwide gross')
plt.xlim(left = 0)
plt.ylim(bottom = 0)
plt.show()
```

###  Evaluating the model

```python
reg.score(X,y)
```
