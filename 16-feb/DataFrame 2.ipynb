### filtering with many conditions

### and 

```python
mask1 = titanic.sex == 'male'
```

```python
mask1.head()
```

```python
mask2 = titanic.age > 14
mask1.head()
```

```python
(mask1 & mask2).head()
```

```python
male_surv = titanic.loc[mask1 & mask2, ['survived','pclass','sex','age']]
male_surv.head()
```

```python
male_surv.info()
```

```python
male_surv.describe()
```

### or

```python
mask1 = titanic.sex == 'female'
mask1.head()
```

```python
mask2 = titanic.age < 14
mask2.head()
```

```python
(mask1 | mask2).head()
```

```python
titanic[mask1 | mask2 ]
```

```python
wom_or_chi = titanic.loc[mask1 | mask2]
```

```python
wom_or_chi
```

```python
wom_or_chi.info()
```

```python
wom_or_chi.describe()
```

### between(), isin() , ~

```python
summer = pd.read_csv('summer.csv')
```

```python
summer.head()
```

```python
og_1988 = summer.loc[summer.Year == 1988]
```

```python
og_1988.head()
```

```python
og_1988.info()
```

```python
og_since1992 = summer.loc[summer.Year >= 1992]
```

```python
og_since1992.head()
```

```python
summer.Year.between(1960,1969).head()
```

```python
og_60s = summer.loc[summer.Year.between(1960,1969,inclusive = True)]
```

```python
og_60s.head()
```

```python
my_fav_games = [1972,1996]
```

```python
summer.Year.isin(my_fav_games).head()
```

```python
og_72_96 = summer.loc[summer.Year.isin(my_fav_games)]
```

```python
og_72_96.head()
```

```python
og_not_72_96 = summer.loc[~summer.Year.isin(my_fav_games)]
```

```python
og_not_72_96
```

```python
og_not_72_96.Year.unique()
```
